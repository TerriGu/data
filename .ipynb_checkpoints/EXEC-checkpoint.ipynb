<h1><center>2017 Index of Economic Freedom<h1><center>

<h2>Introduction<h2>

According to the Heritage Foundation - a conservative American think tank, partner of Wall Street Journal, publisher of the index of econoimc freedom, the index ranks and scores countries based on 12 facotors and numerous measurements: property rights, judicial effectiveness, government integrity, tax burden, government spending, fiscal health, business freedom, labor freedom, monetary freedom, trade freedom, investment freedom and financial freedom. These are compiled into a single score, according to which countries are ranked from most (highest score) to least free.

This report presents a simple analysis of the index of economic freedom in 2017.

<h2>Dataset<h2>

The dataset contains information about 26 measurements of the index of economic freedom for 179 countries in 2017. 3 measurements - GDP growth rate, inflation, and financial freedom - are chosen and analyzed in this report. GDP growth rate measures how fast a country's economy is going. Inflation measures the purchasing power of a country's currency. Financial freedom measures a country and its economy's banking efficiency and independence from goverment control.

The table below is a sample of the three measurements for certain countries.

```python
import pandas as pd
```

```python
#import and read data
dataFile="https://github.com/TerriGu/data/raw/master/index2017.xlsx"
index2017=pd.read_excel(dataFile)
measures=['Country Name','GDP Growth Rate','Inflation Perc','Financial Freedom']
indexm2017=pd.DataFrame(index2017,columns=measures)
indexm2017.head(21)
```

<h2>Descriptive Analysis<h2>

```python
#determine datatype and clear missing data
index2017.dtypes
index_2017=pd.read_excel(dataFile,na_values='')
index_2017.dtypes
```

<h3>1. GDP Growth Rate<h3>

As the table below shows, the average GDP growth rate in 2017 was 2.56% with a standard deviation of 3.86%. Sierra Leon had the smallest GDP growth rate at -21.5%, while Ethiopia's GDP grew the fastest at 10.2%. The 25%, 50%, and 75% quantile were respectively 1.35%, 3.00%, and 4.3%.

```python
# explore the first variable: GDP Growth Rate
index_2017["GDP Growth Rate"].describe()
```

```python
# packages needed to plot the variables
from scipy.stats import norm
import seaborn as sns
import matplotlib.pyplot as plt
```

Plotted in the histogram below, GDP growth rate in 2017 was skewed to the left, with a skewness of -3.00 and a kurtness of 15.56. The blue dashed line shows the mean, while the red dashed line shows the median. The four extreme values that possibly caused the left skewness are dotted in the boxplot.

```python
# plot the variable "GDP Growth Rate" in histogram and box
meanGDP=index_2017['GDP Growth Rate'].mean()
mdGDP=index_2017['GDP Growth Rate'].median()

sns.distplot(index_2017['GDP Growth Rate'].dropna())
plt.title('GDP Growth Rate (%) 2017')
plt.axvline(meanGDP, color='b', linestyle='dashed', linewidth=1,label='mean')
plt.axvline(mdGDP, color='r', linestyle='dashed', linewidth=1,label='median')
plt.show()
```

```python
index_2017["GDP Growth Rate"].skew()
```

```python
index_2017["GDP Growth Rate"].kurt()
```

```python
plot,dataIndex=index_2017['GDP Growth Rate'].plot.box(vert=False,return_type='both')
plt.show()
```

<h3>3. Financial Freedom<h3>

Slightly belwo the median at 50, the average score of financial freedom in 2017 was 48.44 with a standard deviation of 19.10. Afghanistan was least financially free with a score of 0, while Hong Kong had the highest score of 90. Less than 25% of the countries had a score higher than 60.

```python
#explore the third variable: Financial Freedom
index_2017["Financial Freedom"].describe()
```

Financial freedom was alomost normally distributed without outliers. It had a skewness of -0.08 and a kurtness of -0.54.

```python
index_2017["Financial Freedom"].skew()
```

```python
index_2017["Financial Freedom"].kurt()
```

```python
# plot the variable "Financial Freedom" in histogram and box
meanfr=index_2017['Financial Freedom'].mean()
mdfr=index_2017['Financial Freedom'].median()

sns.distplot(index_2017['Financial Freedom'].dropna())
plt.title('Financial Freedom (score) 2017')
plt.axvline(meanfr, color='b', linestyle='dashed', linewidth=1,label='mean')
plt.axvline(mdfr, color='r', linestyle='dashed', linewidth=1,label='median')
plt.show()
```

<h2>Geospacial Analysis<h2>

```python
# load package needed for map visualization
import geopandas as gpd
```

```python
# import and read map file
Map='https://github.com/TerriGu/data/raw/master/ne_110m_admin_0_countries.zip'
```

```python
WorldMap = gpd.GeoDataFrame.from_file(Map)
```

```python
#merge!!!
WM=WorldMap.merge(index_2017, left_on='SOVEREIGNT', right_on='Country Name')
```

The geospacil analysis visualizes countries' financial performance by plotting the three measurements into a world map. Because of missing values in the map file, only 155 countries were successfully plotted.

```python
#before merge, how many rows do I have?
WorldMap.shape[0]
```

```python
#after merge, how many rows do I have?
WM.shape[0]
#and I know I'll be missing 26 countries (177-151)
```

<h3>1. GDP Growth Rate<h3>

The fastest growth appeared in Africa, Asia, and South America. GDP growth rate varied greatly among African and Asian countries. It was relatively alike among countries in North America, but differed in South America. None of the European countries had a GDP growth rate higher than 4.5%, which was the 80% quantile. 

```python
import matplotlib.pyplot as plt

base = WorldMap.plot(color='gray',figsize=(20,13))

topLayer=WM.plot(ax=base,column='GDP Growth Rate',cmap='YlGnBu',scheme='Quantiles',k=5,legend=True,
                  legend_kwds={'loc': 3,'title':'GDP Growth Rate (%): \n (gray missing)'})
topLayer.set_title('2017 Country GDP Growth Rate', 
                   color='black',fontdict={'fontsize':30})
leg = topLayer.get_legend()
plt.setp(leg.get_title(), multialignment='center')
plt.show()
```

<h3>2. Inflation<h3>

Inflaction showed a similar picture, compared with GDP growth rate, except that the highest inflation rate was in Europe instead of Asia. It still varied greatly among African and Asian countries. While most European countries had comparably high GDP growth rate, their inflation rate was comparably low.

```python
base = WorldMap.plot(color='gray',figsize=(20,13))

topLayer=WM.plot(ax=base,column='Inflation Perc',cmap='RdPu',scheme='Quantiles',k=5,legend=True,
                  legend_kwds={'loc': 3,'title':'Inflation (%) : \n (gray missing)'})
topLayer.set_title('2017 Country Inflation', 
                   color='black',fontdict={'fontsize':30})
leg = topLayer.get_legend()
plt.setp(leg.get_title(), multialignment='center')
plt.show()
```

<h3>3. Financial Freedom<h3>

Developed countries generally performed better than developing countries in terms of financial freedom. The highest scores appeared in Australia, North America, and Europe. Most Asian, African, and South American countries had a score lower than 50.

```python
base = WorldMap.plot(color='gray',figsize=(20,13))

topLayer=WM.plot(ax=base,column='Financial Freedom',cmap='YlOrRd',scheme='Quantiles',k=5,legend=True,
                  legend_kwds={'loc': 3,'title':'Financial Freedom (score) : \n (gray missing)'})
topLayer.set_title('2017 Country Financial Freedom', 
                   color='black',fontdict={'fontsize':30})
leg = topLayer.get_legend()
plt.setp(leg.get_title(), multialignment='center')
plt.show()
```
