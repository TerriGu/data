```python
# load package needed for map visualization
import geopandas as gpd
```

```python
# import and read map file
Map='https://github.com/TerriGu/data/raw/master/ne_110m_admin_0_countries.zip'
```

```python
WorldMap = gpd.GeoDataFrame.from_file(Map)
```

```python
WorldMap.head()
```

```python
WorldMap.dtypes
```

```python
# the variable used for merging will be "country," so check it out in the map file first (the name is SOVEREIGNT)
WorldMap['SOVEREIGNT']
```

```python
# then check it out in the data file (the name is Country Name)
index_2017['Country Name']
```

```python
#merge!!!
WM=WorldMap.merge(index_2017, left_on='SOVEREIGNT', right_on='Country Name')
```

```python
#before merge, how many rows do I have?
WorldMap.shape[0]
```

```python
#after merge, how many rows do I have?
WM.shape[0]
#and I know I'll be missing 26 countries (177-151)
```

```python
#plot to see what countries/areas I'm missing
%matplotlib inline
base = WorldMap.plot(color='gray',figsize=(20,13))

WM.plot(ax=base, color='pink')
```

```python
# plot the first variable "GDP growth rate" in map
WM['GDP Growth Rate'].describe()
```

```python
import matplotlib.pyplot as plt

base = WorldMap.plot(color='gray',figsize=(20,13))

topLayer=WM.plot(ax=base,column='GDP Growth Rate',cmap='YlGnBu',scheme='Quantiles',k=5,legend=True,
                  legend_kwds={'loc': 3,'title':'GDP Growth Rate (%): \n (gray missing)'})
topLayer.set_title('2017 Country GDP Growth Rate', 
                   color='black',fontdict={'fontsize':30})
leg = topLayer.get_legend()
plt.setp(leg.get_title(), multialignment='center')
plt.show()
```

```python
# plot the second variable "Financial Freedom" in map
WM['Financial Freedom'].describe(include='all')
```

```python
base = WorldMap.plot(color='gray',figsize=(20,13))

topLayer=WM.plot(ax=base,column='Financial Freedom',cmap='YlOrRd',scheme='Quantiles',k=5,legend=True,
                  legend_kwds={'loc': 3,'title':'Financial Freedom (score) : \n (gray missing)'})
topLayer.set_title('2017 Country Financial Freedom', 
                   color='black',fontdict={'fontsize':30})
leg = topLayer.get_legend()
plt.setp(leg.get_title(), multialignment='center')
plt.show()
```

```python
# plot the third variable "Inflaction Percentage" in map
WM['Inflation Perc'].describe(include='all')
```

```python
base = WorldMap.plot(color='gray',figsize=(20,13))

topLayer=WM.plot(ax=base,column='Inflation Perc',cmap='RdPu',scheme='Quantiles',k=5,legend=True,
                  legend_kwds={'loc': 3,'title':'Inflation (%) : \n (gray missing)'})
topLayer.set_title('2017 Country Inflation', 
                   color='black',fontdict={'fontsize':30})
leg = topLayer.get_legend()
plt.setp(leg.get_title(), multialignment='center')
plt.show()
```
